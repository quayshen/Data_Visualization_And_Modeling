```python
import pandas as pd

# Load dataset
df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

# Preview
print(df.columns)
df.head()

```

```python
df['bpm'].plot(kind='hist', bins=30, title='BPM Distribution', ylabel='Frequency')
print("\nBPM Summary:")
print(df['bpm'].describe())
```

```python
df['key'].value_counts().plot(kind='bar', title='Key Distribution')
print("\nMost Common Key:")
print(df['key'].mode())
```

```python
df['mode'].value_counts().plot(kind='bar', title='Mode Distribution')
print("\nMost Common Mode:")
print(df['mode'].mode())
```

```python
df['danceability_%'].plot(kind='hist', bins=25, title='Danceability Distribution')
print("\nDanceability Summary:")
print(df['danceability_%'].describe())
```

```python
df['energy_%'].plot(kind='hist', bins=25, title='Energy Distribution')
print("\nEnergy Summary:")
print(df['energy_%'].describe())
```

```python
df['speechiness_%'].plot(kind='hist', bins=25, title='Speechiness Distribution')
print("\nSpeechiness Summary:")
print(df['speechiness_%'].describe())
```

```python
df['acousticness_%'].plot(kind='hist', bins=25, title='Acousticness Distribution')
print("\nAcousticness Summary:")
print(df['acousticness_%'].describe())
```

```python
hit_songs = df.loc[
    (df['bpm'] >= 120) & (df['bpm'] <= 140) &
    (df['key'] == 'A') &
    (df['mode'] == 'Minor') &
    (df['danceability_%'] >= 60) & (df['danceability_%'] <= 85) &
    (df['energy_%'] >= 50) & (df['energy_%'] <= 80) &
    (df['speechiness_%'] < 10) &
    (df['acousticness_%'] <= 40)
]

hit_songs[['track_name', 'artist(s)_name', 'bpm', 'key', 'mode', 
           'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%', 
           'streams']].head(10)
```

```python
popular_hits = hit_songs.sort_values(by='streams', ascending=False)
popular_hits[['track_name', 'artist(s)_name', 'streams']].head(10)
```

```python
song = df[df['track_name'].str.contains('Blinding Lights', case=False, na=False)]
song[['track_name', 'artist(s)_name', 'bpm', 'key', 'mode', 
      'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%', 
      'streams']]
```

Based on descriptive statistics, the most common successful song profile includes 120–140 BPM, minor keys like A Minor, medium–high danceability and energy, low speechiness, and low acousticness. Filtering the dataset for these traits returns songs with millions or even billions of streams, validating these as strong predictors of hit songs.
